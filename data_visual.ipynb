```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#charts :line, histogram, scatter plot, pie chrt
x = np.arange(1,7)
y = x**2
z = x**3
print(x)
print(y)
print(z)
```

```python
for i in zip(x,y):
  print(i)
```

```python
#1 line chart
#used when we have sewuential data (gradually increasing or decreasing)
#syntax: plt.plot(x,y)
# creates a 2d line plot
#customization: xlabel, ylabel, color, fontsize
#marker: o, x, *, d, +
plt.plot(x, y, color = "green", label = "square", linestyle="dashed", linewidth = 3, marker = "d", mfc = "lightgrey", mec = "black")
plt.plot(x,z, color = "orange", label = "cubes", marker = "o", mfc = "lightgrey", mec = "black")
plt.xlabel("vals from 1 to 6", color = "blue", fontsize = 13)
plt.ylabel("functions of x", color = "blue", fontsize = 13)
plt.title("sqr and cube vals", color = "red", fontsize = 20)
plt.legend(facecolor = "lightyellow", shadow = True) #creates a container to store labels
plt.grid()
plt.show()
```

```python
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/sales_data_boolean_filtering.csv")
df
```

```python
#total sales by region
```

```python
#group by region, and then visualize each group
region = df.groupby("Region")["TotalSales"].sum()
region = pd.DataFrame(region)
region.reset_index(inplace = True)
plt.plot(region['Region'], region['TotalSales'], marker = "o", mfc = "lightgrey", mec = "black", linewidth = 2.5)
plt.xlabel("Regions", color = "red")
plt.ylabel("Total Sales (Cr)", color = "red")
plt.title("total sales by region", fontsize = 15)
plt.show()
```

```python
#total sales by product
#which product is highly sold
#step 1 unique vals in product
product = df.groupby("Product")["TotalSales"].sum()
product = pd.DataFrame(product)
product.reset_index(inplace = True)
plt.plot(product['Product'], product['TotalSales'], marker = "o", mfc = "lightgrey", mec = "black", linewidth = 2.5)
plt.xlabel("Products", color = "red")
plt.ylabel("Total Sales (Cr)", color = "red")
plt.title("total sales by Product", fontsize = 15)
plt.xticks(rotation = 90)
plt.show()
```

```python
#total sales by category
#which category is highly sold
df.Category.unique()
#step 1 unique vals in category
category = df.groupby("Category")["TotalSales"].sum()
category = pd.DataFrame(category)
category.reset_index(inplace = True)
plt.plot(category['Category'], category['TotalSales'], marker = "o", mfc = "lightgrey", mec = "black", linewidth = 2.5)
plt.xlabel("categories", color = "red")
plt.ylabel("Total Sales (Cr)", color = "red")
plt.title("total sales by category", fontsize = 15)
# plt.xticks(rotation = 90) <- to tilt values
plt.show()
```

```python
#2 subplots:
#to create multple plots
#syntax: fig, axes = plt.subplots(nrows = val, ncols= val)
#axes = all created grids (grid addresses)
#fig = store and access inbuilt methods
x = np.arange(1,7)
y = x**2
z = x**3

fig, axes = plt.subplots(nrows = 2, ncols= 2, figsize =(10, 8), facecolor = "lightyellow") # figsize=(width , height)
# axes[rowidx, colidx].plot(x,y)
fig.suptitle("SAMPLE")
axes[0, 0].plot(x,y, marker = "*")
axes[0, 0].set_xlabel("x", color = "blue")
axes[0, 0].set_ylabel("y", color = "blue")
axes[0, 0].set_title("x vs y", color = "red")
axes[0, 1].plot(x,z, color = "red", marker = "*")
axes[0, 1].set_xlabel("x", color = "blue")
axes[0, 1].set_ylabel("z", color = "blue")
axes[0, 1].set_title("x vs z", color = "red")
axes[1, 0].plot(z,y, color = "green", marker = "*")
axes[1, 0].set_xlabel("z", color = "blue")
axes[1, 0].set_ylabel("y", color = "blue")
axes[1, 0].set_title("z vs y", color = "red")
axes[1, 1].plot(y,z, color = "orange", marker = "*")
axes[1, 1].set_xlabel("y", color = "blue")
axes[1, 1].set_ylabel("z", color = "blue")
axes[1, 1].set_title("y vs z", color = "red")


plt.tight_layout()
```

```python
#group by region, and then visualize each group

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,5), facecolor = "lightblue")
fig.suptitle("TOTAL SALES")

region = df.groupby("Region")["TotalSales"].sum()
region = pd.DataFrame(region)
region.reset_index(inplace = True)
axes[0].plot(region['Region'], region['TotalSales'], color = "orange", marker = "d", mfc = "lightgrey", mec = "black", linewidth = 2.5)
axes[0].set_xlabel("Regions", color = "red")
axes[0].set_ylabel("Total Sales (Cr)", color = "red")
axes[0].set_title("total sales by region", fontsize = 15)
axes[0].grid()

#total sales by category
#which category is highly sold
#step 1 unique vals in category
category = df.groupby("Category")["TotalSales"].sum()
category = pd.DataFrame(category)
category.reset_index(inplace = True)
axes[1].plot(category['Category'], category['TotalSales'], color = "red", marker = "d", mfc = "lightgrey", mec = "black", linewidth = 2.5)
axes[1].set_xlabel("categories", color = "red")
axes[1].set_ylabel("Total Sales (Cr)", color = "red")
axes[1].set_title("total sales by category", fontsize = 15)
axes[1].grid()
# plt.xticks(rotation = 90) <- to tilt values

plt.tight_layout()
plt.show()
```

```python
#3 pie chart
#syntax: plt.pie(data, labels = [], autopct = "%0.2f%%", colors = [], shadow = True/false, explode = ())
#to rep data in terms of percentage
#also for categories
#complete pie chart is of 100%

l = ["python", "ai", "ml", "dl", "testing"]
data = [56, 85, 34, 22, 40]

plt.pie(data, labels = l, autopct = "%0.0f%%", explode = (0,0.1,0.2,0,0.3), shadow = True)
plt.title("student data")
plt.show()
```

## seaborn
1. histplot
2. pairplot
3. heatmap

```python
import seaborn as sns
```

```python
#1 histplot:
# aka histogram
# syntax: sns.histplot(x="colname", y = "colname", data = df, hue = "cat_col")

# works for both categorical and numerical columns
# to understand distribution of data based on one column
# distribution: finding minimum, maximum, and repeated (spread of values)
#used for counting purposes

df.head()
# sns.histplot(y = "type", data = df) # horizontal histplot
sns.histplot(x = "type", data = df, hue= "type") # vertical histplot
```

```python
df["quality"].unique()
# sns.histplot(x="quality", data = df, hue = "quality", palette = "inferno")
# # palette names: inferno, Set2, viridis
sns.histplot(x="density", data = df, kde=True) #kernel density estimation 3.2-3.2: got repeated 350 times
plt.title("density distribution")
plt.grid()
```

```python
#2 pairplot:
#syntax: sns.pairplot(df)
#to understand distribution based on multiple columns
#will create many subplots by default
sns.pairplot(df.iloc[:, 1:3])
# df.iloc[:, ]
```

```python
#3 heatmap():
#syntax: sns.heatmap(data, annot= True, cmap = "str") #cmap= magma, coolwarm, rainbow
#to plot huge data (when we have dataframes)
#to plot multiple columns at the same time
# pivot table: to apply multiple aggr functions on multiple columns at the same time
# syntax: df.pivot_table(values=[col1,col2...],
#         index = "cat_col", aggfunc = ["fname1"...]) # which functions to perform

cols = ["fixed acidity",	"volatile acidity",	"citric acid",	"residual sugar",	"chlorides",	"free sulfur dioxide"]
data = df.pivot_table(values = cols, index = "quality", aggfunc = ["mean"])

#use heatmap
sns.heatmap(data, annot= True, cmap = "coolwarm")
```
